# ingresos totales por tienda :

```python

import pandas as pd


urls = {
    "tienda": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Diccionario para almacenar los ingresos totales de cada tienda
ingresos_totales = {}

# Calcula el ingreso total para cada tienda

for tienda_name, url in urls.items():
    # Cargar los datos desde el archivo CSV

    df = pd.read_csv(url)

    # Calcula la suma de la columna 'Precio'

    ingresos_totales[tienda_name] = df['Precio'].sum()

# Muestra los ingresos totales de cada tienda

for tienda_name, ingreso_total in ingresos_totales.items():
    print(f"Ingresos totales de {tienda_name}: ${ingreso_total:,.2f}")
```

# categoría más populares por tienda

```python

import pandas as pd


urls = {
    "tienda": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

tiendas_data = {}


for tienda_name, url in urls.items():
    tiendas_data[tienda_name] = pd.read_csv(url)

# Función que calcula las categorías más populares de una tienda 5

def get_top_categories(df, top_n=5):
    category_counts = df.groupby('Categoría del Producto').size().reset_index(name='Cantidad_Ventas')
    top_categories = category_counts.sort_values(by='Cantidad_Ventas', ascending=False).head(top_n)
    return top_categories

# Calcula y muestra las categorías más populares para cada tienda

for tienda_name, df in tiendas_data.items():
    top_categories = get_top_categories(df)
    print(f"Categorías más populares de {tienda_name}:")
    print(top_categories)
    print("\n")
```

# calificación promedio por tienda

```python

import pandas as pd


urls = {
    "tienda": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

tiendas_data = {}


for tienda_name, url in urls.items():
    tiendas_data[tienda_name] = pd.read_csv(url)

# Calcula y mostra la calificación total y promedio por tienda

for tienda_name, df in tiendas_data.items():
    calificacion_total = df['Calificación'].sum()
    calificacion_promedio = df['Calificación'].mean()

    print(f"Tienda: {tienda_name}")
    print(f"Calificación total: {calificacion_total:.2f}")
    print(f"Calificación promedio: {calificacion_promedio:.2f}\n")

    # Calcula la calificación promedio por categoría

    calificacion_promedio_por_categoria = df.groupby('Categoría del Producto')['Calificación'].mean().reset_index(name='Calificación Promedio')

    print(f"Calificación promedio por categoría en {tienda_name}:")
    print(calificacion_promedio_por_categoria)
    print("\n")
```

# productos más y menos vendidos y el costo promedio por tienda

```python

import pandas as pd


urls = {
    "tienda": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}


tiendas_data = {}

# datos de cada tienda

for tienda_name, url in urls.items():
    tiendas_data[tienda_name] = pd.read_csv(url)

# Función para obtener los 3 productos más y menos vendidos 3

def get_top_and_bottom_products(df, top_n=3, bottom_n=3):
    product_counts = df.groupby('Producto').size().reset_index(name='Cantidad_Ventas')
    top_products = product_counts.sort_values(by='Cantidad_Ventas', ascending=False).head(top_n)
    bottom_products = product_counts.sort_values(by='Cantidad_Ventas', ascending=True).head(bottom_n)
    return top_products, bottom_products

# Calcula y muestra los resultados para cada tienda

for tienda_name, df in tiendas_data.items():
    top_products, bottom_products = get_top_and_bottom_products(df)

    print(f"Tienda: {tienda_name}")

    print("\n3 Productos más vendidos:")
    print(top_products)

    print("\n3 Productos menos vendidos:")
    print(bottom_products)

    # Calcula  el promedio del costo de envío

    promedio_costo_envio = df['Costo de envío'].mean()
    print(f"\nPromedio del costo de envío: ${promedio_costo_envio:,.2f}")

    print("\n")
```

# GRÁFICOS

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown


urls = {
    "tienda": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}


tiendas_data = {}


for tienda_name, url in urls.items():
    tiendas_data[tienda_name] = pd.read_csv(url)

# Función para obtener los 3 productos más y menos vendidos

def get_top_and_bottom_products(df, top_n=3, bottom_n=3):
    product_counts = df.groupby('Producto').size().reset_index(name='Cantidad_Ventas')
    top_products = product_counts.sort_values(by='Cantidad_Ventas', ascending=False).head(top_n)
    bottom_products = product_counts.sort_values(by='Cantidad_Ventas', ascending=True).head(bottom_n)
    return top_products, bottom_products

# Calcula y muestra los resultados para cada tienda

tiendas_resultados = {}

for tienda_name, df in tiendas_data.items():
    top_products, bottom_products = get_top_and_bottom_products(df)
    calificacion_total = df['Calificación'].sum()
    calificacion_promedio = df['Calificación'].mean()
    promedio_cuotas = df['Cantidad de cuotas'].mean()
    promedio_costo_envio = df['Costo de envío'].mean()

    tiendas_resultados[tienda_name] = {
        'top_products': top_products,
        'bottom_products': bottom_products,
        'calificacion_total': calificacion_total,
        'calificacion_promedio': calificacion_promedio,
        'promedio_cuotas': promedio_cuotas,
        'promedio_costo_envio': promedio_costo_envio
    }

# Gráfico de barras para mostrar las ventas totales por tienda

plt.figure(figsize=(10, 6))
tiendas_ventas_totales = {tienda_name: len(df) for tienda_name, df in tiendas_data.items()}
sns.barplot(x=list(tiendas_ventas_totales.keys()), y=list(tiendas_ventas_totales.values()), palette='viridis')
plt.title('Ventas totales por tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de ventas')
plt.ylim(0, max(tiendas_ventas_totales.values()) + 10)
plt.show()

display(Markdown("### Ventas totales por tienda"))
display(Markdown("Este gráfico muestra la cantidad total de ventas realizadas por cada tienda."))

# Gráficos de pastel para las categorías más vendidas por tienda

for tienda_name, df in tiendas_data.items():
    category_counts = df.groupby('Categoría del Producto').size().reset_index(name='Cantidad_Ventas')
    top_categories = category_counts.sort_values(by='Cantidad_Ventas', ascending=False).head(5)

    plt.figure(figsize=(8, 8))
    plt.pie(top_categories['Cantidad_Ventas'], labels=top_categories['Categoría del Producto'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
    plt.title(f'Categorías más vendidas en {tienda_name}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

    display(Markdown(f"### Categorías más vendidas en {tienda_name}"))
    display(Markdown("Este gráfico muestra las categorías más vendidas en la tienda."))

# Gráfico de línea para mostrar el promedio total de las calificaciones totales de cada tienda

plt.figure(figsize=(10, 6))
tiendas_calificaciones_promedio = {tienda_name: result['calificacion_promedio'] for tienda_name, result in tiendas_resultados.items()}
sns.lineplot(x=list(tiendas_calificaciones_promedio.keys()), y=list(tiendas_calificaciones_promedio.values()), marker='o', color='blue')
plt.title('Promedio total de las calificaciones por tienda')
plt.xlabel('Tienda')
plt.ylabel('Promedio de calificación')
plt.ylim(0, max(tiendas_calificaciones_promedio.values()) + 1)
plt.grid(True)
plt.show()

display(Markdown("### Promedio total de las calificaciones por tienda"))

display(Markdown("Este gráfico muestra el promedio total de las calificaciones dadas por los clientes en cada tienda."))

# Gráfico de dispersión para mostrar el promedio total del costo de envío de cada tienda

plt.figure(figsize=(10, 6))
tiendas_costo_envio_promedio = {tienda_name: result['promedio_costo_envio'] for tienda_name, result in tiendas_resultados.items()}
sns.scatterplot(x=list(tiendas_costo_envio_promedio.keys()), y=list(tiendas_costo_envio_promedio.values()), s=100, color='green')
plt.title('Promedio total del costo de envío por tienda')
plt.xlabel('Tienda')
plt.ylabel('Promedio del costo de envío')
plt.ylim(0, max(tiendas_costo_envio_promedio.values()) + 1000)
plt.grid(True)
plt.show()

display(Markdown("### Promedio total del costo de envío por tienda"))
display(Markdown("Este gráfico muestra el promedio total del costo de envío para cada tienda."))
```
